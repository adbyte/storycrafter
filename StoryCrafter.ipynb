StoryCrafter, Short Story Generator for Kids.

Fun and educational Generative AI Tool,designed especially for kids, It lets users input a theme a character and then uses a pre trained GPT-2 model from Hugging Face to genrrate a unique and engagin short story.

App is built using -:

Transformers lubrary (for gpt-2 story generation)
Gradio (for interactive web ui)
Google collab (for backend development and depolyment)

```python
from transformers import pipeline, set_seed
story_generator = pipeline("text-generation", model="gpt2")
set_seed(42)
```

```python
def generate_story(theme, character, location):
    prompt = f"Once upon a time in {location}, there lived a {character} who loved {theme}. One day,"
    story = story_generator(prompt, max_length=200, num_return_sequences=1)[0]['generated_text']
    return story
```

```python
import gradio as gr
iface = gr.Interface(
    fn=generate_story,
    inputs=[
        gr.Textbox(label="Theme (eg, Adventure, Kindness, Magic)" ),
        gr.Textbox(label ="Main character (eg, unicorm, fairy, demon)"),
        gr.Textbox(label ="Location (eg, enchanted forest, outer space)")
    ],
    outputs=gr.Textbox(label="Generated Story"),
    title="StoryCrafter - Short Story Generator for Kids",
    description="Enter a theme, character, and location to generate a unique and engaging short story."
)
```

```python
iface.launch(share = True)
```
